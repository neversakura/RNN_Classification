```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

testY=np.load('TestY.npy')
predictY=np.load('PredictY.npy')
```

```python
timeSeriesLen=testY.shape[1]
x=np.arange(timeSeriesLen)
testYsample=testY[0,:,:]
predictSample=predictY[0,:,:]
plt.clf()
plt.figure(figsize = (9,7))
gs=gridspec.GridSpec(6,1)
gs.update(hspace=0.0)
ax={}
for i in range(6):
    ax[i]=plt.subplot(gs[i])

lineOne, = ax[0].plot(x,testYsample[:,0],'r')
lineTwo, = ax[0].plot(x,testYsample[:,3],'y')
lineThree, =ax[0].plot(x,predictSample[:,0],'r--')
lineFour, =ax[0].plot(x,predictSample[:,3],'y--')
ax[0].set_ylim(0.605,0.65)
ax[0].set_yticks(np.arange(0.615,0.64,0.01))
ax[0].legend([lineOne, lineTwo, lineThree, lineFour], ['R1', 'R4', 'R1 Prediction', 'R2 Prediction'],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=10)
ax[0].set_xlim(0,540)

lineOne, = ax[1].plot(x,testYsample[:,1],'b')
lineTwo, = ax[1].plot(x,testYsample[:,2],'g')
lineThree, = ax[1].plot(x,predictSample[:,1],'b--')
lineFour, = ax[1].plot(x,predictSample[:,2],'g--')
ax[1].set_ylim(0.61,0.66)
ax[1].set_yticks(np.arange(0.62,0.65,0.02))
ax[1].legend([lineOne, lineTwo, lineThree, lineFour], ['R2', 'R3', 'R2 Prediction', 'R3 Prediction'],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=10)
ax[1].set_xlim(0,540)

lineOne, = ax[2].plot(x,testYsample[:,4],'r')
lineTwo, = ax[2].plot(x,testYsample[:,5],'y')
lineThree, = ax[2].plot(x,predictSample[:,4],'r--')
lineFour, = ax[2].plot(x,predictSample[:,5],'y--')
ax[2].set_ylim(-0.97,-0.6)
ax[2].set_yticks(np.arange(-0.65,-0.97,-0.2))
ax[2].legend([lineOne, lineTwo, lineThree, lineFour], ['R5', 'R6', 'R5 Prediction', 'R6 Prediction'],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=10)
ax[2].set_xlim(0,540)

lineOne, = ax[3].plot(x,testYsample[:,6],'g')
lineTwo, = ax[3].plot(x,testYsample[:,7],'b')
lineThree, = ax[3].plot(x,predictSample[:,6],'g--')
lineFour, = ax[3].plot(x,predictSample[:,7],'b--')
ax[3].set_ylim(-0.94,-0.83)
ax[3].set_yticks(np.arange(-0.87,-0.9,-0.03))
ax[3].set_xlim(0,540)
ax[3].legend([lineOne, lineTwo, lineThree, lineFour], ['R7', 'R8', 'R7 Prediction', 'R8 Prediction'],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=10)

lineOne, = ax[4].plot(x,testYsample[:,8],'c')
lineTwo, = ax[4].plot(x,predictSample[:,8],'c--')
ax[4].set_yticks(np.arange(0,0.18,0.08))
ax[4].set_xlim(0,540)
ax[4].legend([lineOne, lineTwo], ['Temperature', 'Temp Prediction'], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=10)

lineOne, = ax[5].plot(x,testYsample[:,9],'m')
lineTwo, = ax[5].plot(x,predictSample[:,9],'m--')
ax[5].set_yticks(np.arange(-0.16,-0.30,-0.1))
ax[5].set_xlim(0,540)
ax[5].legend([lineOne, lineTwo], ['Humidity', 'Humidity Prediction'], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=10)

ax[0].set_title('Background')

plt.show()
```
